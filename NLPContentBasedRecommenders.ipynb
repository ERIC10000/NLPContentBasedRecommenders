```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://msi.martial.co.ke/datasets/news.csv')
df.head()
```

```python
# News Dataset Infomation
df.info()
```

```python
# Out of 3421 News Records, we can take the first 1000 news records
# To save on Computing time.

df = df.head(1000)
df.shape
```

```python
# Extract the Contents of the News Articles
content = df['content']
```

## What are Embeddings in NLP

This is a technique in NLP where words, phrases or entire texts is represented as numerical vectors in a continous space.This will algorithms to work effectively with textual data because they  are represented as numerical vectors.

In Python, to convert the textual into numerical embeddings, we use the sentence-transformers library.

```python
!pip install sentence-transformers
```

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('distilbert-base-nli-stsb-mean-tokens')
```

```python
embeddings = model.encode(content, show_progress_bar=True)
```

```python
# Convert the Embeddings into a DataFrame
embeddings_df = pd.DataFrame(embeddings)
embeddings_df.head()
```

## Cosine Similarity
In Mathematics, it is a measure used to determine how similar two vectors are, regardless of their magnitude. In NLP it is used to measure the similar of vector embedding, to determine phrases, words and sentenses that refer to the same meaning.

```python
from sklearn.metrics.pairwise import cosine_similarity
cos_sim = cosine_similarity(embeddings_df)
cos_sim
```

```python
# Convert the Cosine Similarities in a DataFrame
cos_sim_df = pd.DataFrame(cos_sim)
cos_sim_df.head()
```
